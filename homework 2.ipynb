```python
#giai phuong trình bậc 2:ax^2+bx+c=0
a=float(input('Nhap he so a:'))
b=float(input('Nhap he so b:'))
c=float(input('Nhap he so c:'))
if a==0:
    if b==0:
        if c==0:
            print('phương trinh bac 2 vô số nghiệm')
        else:
            print('phương trinh bậc 2 vô nghiệm')
    else:
        print('phương trinh bậc 2 co nghiem x=',-c/b)
else:
    denta=b**2-4*a*c
    if denta==0:
        print('phương trình bac 2 có nghiem kép là x=',-b/(2*a))
    elif denta<0:
        print('phuong trinh bac hai vô nghiệm')
    else:
        print('phuong trinh co 2 nghiem phan biet là:')
        x1=(-b-denta**0.5)/(2*a)
        x2=(-b+denta**0.5)/(2*a)
        print('x1=',x1,'x2=',x2)
        
```
