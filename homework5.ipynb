Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->
int:

```python
def max_of_two(a,b):
    if a>b:
        return a
    elif b>a:
        return b
    else:
        return a,b
a,b=map(int(input().split())) 
print(max_of_two(a,b))
```

Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def check_prime(n):
    if n<2:
        return False
    for i in range(2 ,int(n**0.5+1)):
        if n%i==0:
            return False 
    return True
n=int(input())
if check_prime(n):
    print("True")
else:
    print("False")
    
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def check_vitri(chuoi,k):
    vitri=0
    if k in chuoi :
        for i in chuoi:
            vitri +=1
            if i==k:
                break
        return vitri    
    else:
        return False
chuoi=list(map(int,input().split()))
k=int(input())
if check_vitri(chuoi,k):
    print(check_vitri(chuoi,k))        
else:
    print(-1)
        
```

Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def hamming_distance(x,y):
    # Bước 1: XOR để tìm các bit khác nhau
    xor_result = x ^ y
    
    # Bước 2: Đếm số lượng bit 1 trong xor_result
    distance = 0
    while xor_result > 0:
        # Kiểm tra bit cuối cùng có phải là 1 hay không
        distance += xor_result & 1
        # Dịch phải 1 bit để kiểm tra bit tiếp theo
        xor_result >>= 1
        
    return distance
x,y=map(int,input().split())
print(f'khoang cach hamming giua 2 so x va y la :{hamming_distance(x,y)}')
 
```

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
def kiem_tra_dang_cau():
    # Nhập hai từ từ bàn phím
    tu_a = input("Nhập từ thứ nhất (ví dụ abca): ")
    tu_b = input("Nhập từ thứ hai (ví dụ zbxz): ")

    # Hàm nhỏ xử lý logic kiểm tra
    def la_isomorphic(s, t):
        # Dictionary để lưu ánh xạ từ s sang t
        mapping = {}
        # Set để theo dõi các ký tự trong t đã được dùng để ánh xạ chưa
        seen_in_t = set()

        for char_s, char_t in zip(s, t):
            if char_s in mapping:
                # Nếu char_s đã có ánh xạ, nó phải khớp với char_t hiện tại
                if mapping[char_s] != char_t:
                    return False
            else:
                # Nếu char_s chưa có ánh xạ, thì char_t cũng phải chưa được dùng
                if char_t in seen_in_t:
                    return False
                
                # Thiết lập mối quan hệ ánh xạ
                mapping[char_s] = char_t
                seen_in_t.add(char_t)
#zip(s, t): Hàm này ghép cặp từng ký tự của hai từ lại với nhau. Ví dụ: zip("abc", "zbx") sẽ tạo ra các cặp ('a', 'z'), ('b', 'b'), ('c', 'x').

#mapping: Giống như một cuốn sổ tay ghi chú. Nếu ta đã nói 'a' biến thành 'z', lần sau gặp 'a' mà thấy nó đòi biến thành 'y', hàm sẽ trả về False ngay.

#seen_in_t: Dùng để tránh trường hợp cả 'a' và 'c' đều đòi biến thành 'z'. Quy tắc là mỗi ký tự chỉ có một "người yêu" duy nhất
        
        return True

    # Kiểm tra điều kiện độ dài (theo giả thiết bài toán là cùng độ dài)
    if len(tu_a) != len(tu_b):
        print("False (Độ dài không bằng nhau)")
    else:
        if la_isomorphic(tu_a, tu_b):
            print("True (Hai từ đẳng cấu)")
        else:
            print("False (Hai từ không đẳng cấu)")

# Chạy chương trình
kiem_tra_dang_cau()


```


Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    if divisor >0 and dividend > divisor:
        q=dividend//divisor
        r=dividend%divisor
        if divisor > r :
            return q,r
    elif divisor < 0 and dividend < divisor :
        q=dividend//divisor
        r=dividend%divisor
        if divisor < r :
            return q,r
    else:
        return False 
dividend,divisor=map(int,input().split())
print(divide_with_remainder(dividend,divisor))

    

```

Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a,b,op):
    dau_ra= None 
    if op == "+":
        dau_ra=a+b
        return dau_ra
    elif op=="-":
        dau_ra=a-b
        return dau_ra
    elif op=="/":
        dau_ra=a/b
    elif op=="*":
        dau_ra=a*b
        return dau_ra
    
a,b=map(int,input().split())
op=input()
if compute(a,b,op):
    print(f"ket qua dau ra cua phep toan la :{compute(a,b,op)}")
else:
    print("False")
```

Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a,b,c):
    if a+b>c and b+c>a and c+a>b:
        return True
    else:
        return False
a,b,c=map(int,input().split())
if is_right_triangle(a,b,c):
    print(is_right_triangle(a,b,c))
```
