"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
def kiemtra_prime(x):
    if x < 2:
        return False
    for i in range(2,int(x**0.5+1)):
        if x%i==0:
            return False
    return True

a,b = map(int,input().split())
so_prime=[]
for x in range(a,b+1):
    if kiemtra_prime(x):
        so_prime.append(x)
total=sum(so_prime)
print(total)
```

ReverseAddPalindrome
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được

```python
def reverse_num(n):
    return int(str(n)[::-1])
def is_palindrome(n):
    return n == reverse_num(n)

n = int(input("Nhập số nguyên dương n: "))
print("bắt đầu với số {n}")
step = 0
while not is_palindrome(n):
    rev = reverse_num(n)
    n_cu = n
    n = n + rev
    step += 1
    print(f"{step}: {n_cu} + {rev} = {n}")

print(f" {n} là số đối xứng sau {step} bước")
```

"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 

```python
def solve_collatz(n):
    # Bộ nhớ đệm để lưu độ dài đã tính: {số: độ_dài}
    memo = {1: 1}
    
    max_length = 0
    best_x = 0

    for i in range(1, n + 1):
        current = i
        path = []
        
        # Tính độ dài dãy Collatz cho i
        while current not in memo:
            path.append(current)
            if current % 2 == 0:
                current //= 2
            else:
                current = 3 * current + 1
        
        # Cập nhật độ dài vào memo cho các số trong path
        temp_length = memo[current]
        for step in reversed(path):
            temp_length += 1
            memo[step] = temp_length
            
        # Kiểm tra xem i có phải là số tốt nhất không
        if memo[i] > max_length:
            max_length = memo[i]
            best_x = i
        elif memo[i] == max_length:
            # Nếu độ dài bằng nhau, chọn số nhỏ nhất
            best_x = min(best_x, i)
            
    return best_x, max_length

# Nhập dữ liệu
try:
    n = int(input("Nhập số nguyên dương n: "))
    if n <= 0:
        print("Vui lòng nhập số dương.")
    else:
        x, L = solve_collatz(n)
        print(f"{x} {L}")
except ValueError:
    print("Dữ liệu nhập không hợp lệ.")
```

Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết

110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220

Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a,b=map(int,input().split())
uoca=[]
uocb=[]
for i in range(1,a):
    if a%i==0:
        uoca.append(i)
    tonga=sum(uoca)
for j in range(1,b):
    if b&j==0:
        uocb.append(j)
    tongb=sum(uocb)
if tonga==b and tongb==a:
    print('this is closepairnumber')
else:
    print("this isnt closepairnumber")
```

[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n.

```python
m,n=map(int,input().split())
uocm=[]
uocn=[]
for i in range(1,m):
    if m%i==0:
        uocm.append(i)
   
for j in range(1,n):
    if n&j==0:
        uocn.append(j)
uoc_chung=[]
for i in uocm:
    if i in uocn:
        uoc_chung.append(i)
print(max(uoc_chung))
```
