```python
# Hàm tính tiền điện theo bậc thang và cộng thêm VAT 8%
def tinh_tien_dien(chi_so_cu, chi_so_moi):
    so_kwh = chi_so_moi - chi_so_cu
    tien = 0
    
    # Bậc 1: 0 - 50 kWh
    if so_kwh <= 50:
        tien = so_kwh * 1984
    elif so_kwh <= 100:
        tien = 50 * 1984 + (so_kwh - 50) * 2050
    elif so_kwh <= 200:
        tien = 50 * 1984 + 50 * 2050 + (so_kwh - 100) * 2380
    elif so_kwh <= 300:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_kwh - 200) * 2998
    elif so_kwh <= 400:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_kwh - 300) * 3350
    else:
        tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (so_kwh - 400) * 3460
    
    # Cộng thêm VAT 8%
    tong_tien = tien * 1.08
    return round(tong_tien)

# Nhập dữ liệu
ho_ten = input("Ten chu ho: ")
chi_so_cu = int(input("Chi so thang truoc: "))
chi_so_moi = int(input("Chi so thang nay: "))

# Tính tiền
tien_phai_tra = tinh_tien_dien(chi_so_cu, chi_so_moi)

# Xuất kết quả
print(f"Ho va ten: {ho_ten}")
print(f"Tien phai tra la: {tien_phai_tra}")
```
