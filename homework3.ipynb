Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý

```python
n=int(input())
if 100 <= n <=999 :
    tram=n//100
    chuc=(n - tram*100)//10
    don=n-tram*100-chuc*10
    so_dao=tram+chuc*10+don*100
    print(f'so dao nguoc la {so_dao}')
else:
    print('vui long nhap dung so co 3 chu so')
```

Viêt chương trình chia một số m cho n, kết quả thu được làm tròn xuống
Viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên

```python
m,n=map(int,input().split())
if m==0 and n!=0:
    print(m/n)
elif m%n==0:
    print(m/n)
elif (m >0 and n >0) or (m <0 and n <0):
    if (m/n - m//n)>= 0.5 :
        result=m//n + 1 
        print(result)
    else:
        result=m//n
        print(result)
elif (m > 0 and n < 0) or ( m<0 and n>0):
    if 0 > (m/n-m//n) >=-0.5:
        result=m//n+1
        print(result)
    else:
        result=m//n
        print(result)
```

tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều",
"Tam giác cân", hay "Tam giác thường"

```python
a,b,c=map(int,input().split())
if a+b>c and b+c>a and c+a>b:
    if a==b and b==c and c==a :
        print('tam giac deu')
    elif a==b or b==c or c==a :
        print('tam giac can')
    elif a**2+b**2==c**2 or b**2+c**2==a**2 or a**2+c**2==b**2:
        print('tam giac vuong')
    else:
        print('tam giac thuong')
else:
    print('ko tao thanh 1 tam giac')
    
```
